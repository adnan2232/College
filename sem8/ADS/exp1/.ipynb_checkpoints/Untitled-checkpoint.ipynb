```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("car.csv")
```

```python
df.head()
```

```python
df = df.applymap(lambda x: x if x!="?" else np.nan)
```

```python
df.head()
```

```python
df.info()
```

```python
np.sum(df.isnull())
```

```python
fig,ax = plt.subplots(1,1,figsize=(10,9))
sns.heatmap(df.corr(),center=0.5,annot=True,vmin=0,ax=ax)
plt.show()
```
